```python
# Read the AnnData file
adata_covid = sc.read('/content/drive/My Drive/Singlecell/2301_annot_patient_integrated.h5ad')
```

```python
# Subsample to 10,000 cells
target_num_cells = 10000
sc.pp.subsample(adata_covid, n_obs=target_num_cells, random_state=42)
```

```python
adata_covid
```

```python
# Define a list of model names
model_names = ['Cells_Fetal_Lung.pkl', 'Immune_All_High.pkl', 'Immune_All_Low.pkl', 'Healthy_COVID19_PBMC.pkl']
# Iterate over each model name
for model_name in model_names:
#Load the model
  model = models.Model.load(model=model_name)
  # Print information about the loaded model
  print(f"Loaded model: {model_name}")
  # Annotate cells using the current model
  predictions = celltypist.annotate(adata_covid, model=model, majority_voting=True)
  # Print predicted labels
  print(f"Predicted labels for {model_name}: {predictions.predicted_labels}")
  # Convert predictions to AnnData object
  adata = predictions.to_adata()
  # Print cell types in the annotated data
  print(f"Cell types in the annotated data for {model_name}: {adata.obs.cell_type}")
  # Perform UMAP analysis
  sc.tl.umap(adata)
  # Plot UMAP
  sc.pl.umap(adata, color=['cell_type', 'predicted_labels', 'majority_voting'], legend_loc='on data')
  # Dotplot analysis
  celltypist.dotplot(predictions, use_as_reference='cell_type', use_as_prediction='majority_voting')
```
